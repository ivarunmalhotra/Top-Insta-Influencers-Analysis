### Data Cleaning

```python
#Converting posts value into numeric format
posts=[]
for i in data['posts']:
    if 'k' in i:
        posts+=[int(float(i.replace('k',''))*1000)]
    else:
        posts+=[int(i)]
data['posts']=posts
```

```python
#Converting followers value into numeric format
followers=[]
for i in data['followers']:
    if 'm' in i:
        followers+=[int(float(i.replace('m',''))*1000000)]
    else:
        followers+=[int(i)]
data['followers']=followers
```

```python
#Converting avg likes value into numeric format
avg_likes=[]
for i in data['avg_likes']:
    if 'm' in i:
        avg_likes+=[int(float(i.replace('m',''))*1000000)]
    elif 'k' in i:
        avg_likes+=[int(float(i.replace('k',''))*1000)]
    else:
        avg_likes+=[int(i)]
data['avg_likes']=avg_likes
```

```python
#Converting total likes value into numeric format
total_likes=[]
for i in data['total_likes']:
    if 'b' in i:
        total_likes+=[int(float(i.replace('b',''))*1000000000)]
    elif 'm' in i:
        total_likes+=[int(float(i.replace('m',''))*1000000)]
    elif 'k' in i:
        total_likes+=[int(float(i.replace('k',''))*1000)]
    else:
        total_likes+=[int(i)]
data['total_likes']=total_likes
```

## Exploratory Data Analysis
### Heatmap

```python
#Extracting numeric column names from the dataset
numeric_column=data.select_dtypes(include=np.number).columns.tolist()
numeric_column
```

```python
corrmat = data[numeric_column].corr()
f, ax = plt.subplots(figsize = (7,5))
sns.heatmap(corrmat, vmax=1, annot=True)
```

### Scatterplot

```python
plt.figure(figsize = (12,10))
sns.pairplot(data[numeric_column], height=3, kind="scatter", diag_kind="kde")
plt.show()
```

### Line Chart

```python
posts = data.groupby('channel_info')['posts'].mean().sort_values().tail(10)

# Plot
plt.figure(figsize=(10,6))
sns.lineplot(x=posts.index, y=posts.values, marker='o', color='blue')
plt.title('Top 10 influencers with highest engagement')
plt.xlabel('influencer')
plt.xticks(rotation=90)
plt.ylabel('posts')
plt.show()
```

### Bar Chart
#### for top 10 influencers

```python
# Plot barcharts for Top 10 influencers by posts, followers, avg likes and total likes
plt.figure(figsize=(12,11))

plt.subplot(2,2,1)
data_sorted = data.sort_values(by='posts', ascending=False)
sns.barplot(data=data_sorted, x=data_sorted.iloc[:10,1], y=data_sorted.iloc[:10,3])
plt.xlabel('Influencers')
plt.xticks(rotation=90)
plt.ylabel('Posts')
plt.title('Top 10 Influencers by Posts')

plt.subplot(2,2,2)
data_sorted = data.sort_values(by='followers', ascending=False)
sns.barplot(data=data_sorted, x=data_sorted.iloc[:10,1], y=data_sorted.iloc[:10,4])
plt.xlabel('Influencers')
plt.xticks(rotation=90)
plt.ylabel('Followers')
plt.title('Top 10 Influencers by Followers')

plt.subplot(2,2,3)
data_sorted = data.sort_values(by='avg_likes', ascending=False)
sns.barplot(data=data_sorted, x=data_sorted.iloc[:10,1], y=data_sorted.iloc[:10,5])
plt.xlabel('Influencers')
plt.xticks(rotation=90)
plt.ylabel('Avg Likes')
plt.title('Top 10 Influencers by Avg Likes')

plt.subplot(2,2,4)
data_sorted = data.sort_values(by='total_likes', ascending=False)
sns.barplot(data=data_sorted, x=data_sorted.iloc[:10,1], y=data_sorted.iloc[:10,8])
plt.xlabel('Influencers')
plt.xticks(rotation=90)
plt.ylabel('Total Likes')
plt.title('Top 10 Influencers by Total Likes')

plt.tight_layout()
plt.show()
```

### Bar Chart showing Countries and their influence scores

```python
#Aranging Country according to the max influence_score in descending order
country_influence=data[['country','influence_score']]
country_influence=country_influence.groupby('country', as_index=False)['influence_score'].max()
country_influence=country_influence.sort_values('influence_score',ascending=False)
country_influence.head()
```

```python
#Bar Chart
plt.figure(figsize = (12,5))
sns.barplot(data=country_influence, x=country_influence.iloc[:, 0], y=country_influence.iloc[:, 1])
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Influence Score')
plt.title('Bar Chart showing Influence Score by Country')
plt.show()
```
